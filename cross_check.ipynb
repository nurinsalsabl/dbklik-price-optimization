```python
import pandas as pd
df_dbklik = pd.read_excel("/content/produk_ada_hpp_dan_tambahan_ts.xlsx")
```

```python
df_dbklik.head()
```

```python
print("HPP=1:",(df_dbklik["HPP (Latest)"] == 1).sum())

df_dbklik = df_dbklik[(df_dbklik["HPP (Latest)"].isna())|(df_dbklik["HPP (Latest)"] > 1)]
```

```python
df_dbklik = df_dbklik[df_dbklik["HPP (Latest)"].notna()]

df_dbklik = df_dbklik[df_dbklik["HPP (Latest)"] > 1]
```

```python
print("Jumlah data:", len(df_dbklik))

print("HPP NaN:",df_dbklik["HPP (Latest)"].isna().sum())
print("HPP = 1:",(df_dbklik["HPP (Latest)"] == 1).sum())
print("HPP <= 1:",(df_dbklik["HPP (Latest)"] <= 1).sum())
```
