# cek coverage setelah normalisasi

```python
import re

def clean_sku(x):

    x = str(x).upper()

    x = re.sub(
        r'-.*$',
        '',
        x
    )

    x = re.sub(
        r'\.P$',
        '',
        x
    )

    return x.strip()

df_final["SKU_BASE"] = (
    df_final["SKU"]
    .apply(clean_sku)
)

df_hpp["SKU_BASE"] = (
    df_hpp["SKU"]
    .apply(clean_sku)
)
```

```python
base_ts = set(
    df_final["SKU_BASE"]
)

base_hpp = set(
    df_hpp["SKU_BASE"]
)

print(
    "Masih tidak ketemu:",
    len(base_ts - base_hpp)
)
```

```python
from rapidfuzz import process, fuzz
import pandas as pd

# Produk yang belum punya HPP
df_hpp_tidak_ada = df_final[
    df_final["HPP (Latest)"].isna()
].copy()

print(df_hpp_tidak_ada.shape)

# List SKU dari portal HPP
sku_hpp_list = (
    df_hpp["SKU"]
    .dropna()
    .astype(str)
    .unique()
)

hasil = []

for idx, row in df_hpp_tidak_ada.iterrows():

    sku_ts = str(row["SKU"])

    kandidat = process.extractOne(
        sku_ts,
        sku_hpp_list,
        scorer=fuzz.ratio
    )

    if kandidat:

        hasil.append([
            row["Nama Produk"],
            sku_ts,
            kandidat[0],      # SKU HPP kandidat
            kandidat[1]       # Similarity score
        ])

df_fuzzy = pd.DataFrame(
    hasil,
    columns=[
        "Nama Produk",
        "SKU_TS",
        "SKU_HPP_Kandidat",
        "Similarity"
    ]
)

# Ambil yang similarity >= 98%
df_fuzzy_98 = df_fuzzy[
    df_fuzzy["Similarity"] >= 60
]

print(df_fuzzy_98.shape)

# df_fuzzy_98.to_excel(
#     "hpp_kandidat_similarity_98.xlsx",
#     index=False
# )
```

```python
# Lihat statistik similarity
print(df_fuzzy["Similarity"].describe())
```

```python
df_fuzzy_98 = df_fuzzy_98.merge(

    df_hpp[
        [
            "SKU",
            "HPP (Latest)"
        ]
    ],

    left_on="SKU_HPP_Kandidat",
    right_on="SKU",
    how="left"

)
```
